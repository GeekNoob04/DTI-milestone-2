```python
import requests
from bs4 import BeautifulSoup
import csv

def scrape_inshorts():
    # URL of the website to scrape
    url = "https://inshorts.com/en/read"

    # Send a GET request to the URL with headers
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"}
    response = requests.get(url, headers=headers)

    # Parse the HTML content
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all the cards containing news articles
    cards = soup.find_all(class_='news-card')

    # Check if any cards were found
    print("Number of cards found:", len(cards))

    # Initialize lists to store data
    headlines = []
    summaries = []
    links = []

    # Loop through each card and extract relevant information
    for card in cards:
        # Extract headline
        headline_elem = card.find(class_='news-card-title')
        if headline_elem:
            headline = headline_elem.find('a').text.strip()
            headlines.append(headline)
            print("Headline:", headline)
        else:
            print("No headline found")

        # Extract summary
        summary_elem = card.find(class_='news-card-content')
        if summary_elem:
            summary = summary_elem.find('div').text.strip()
            summaries.append(summary)
            print("Summary:", summary)
        else:
            print("No summary found")

        # Extract link
        link_elem = card.find(class_='news-card-title')
        if link_elem:
            link = "https://inshorts.com" + link_elem.find('a')['href']
            links.append(link)
            print("Link:", link)
        else:
            print("No link found")

        print()

    # Write the data to a CSV file
    with open('inshorts_articles.csv', 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(["Headline", "Summary", "Link"])
        writer.writerows(zip(headlines, summaries, links))

    print("Scraping completed and data saved to 'inshorts_articles.csv'.")

if __name__ == "__main__":
    scrape_inshorts()
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import csv
import time

def scrape_inshorts():
    # URL of the website to scrape
    url = "https://inshorts.com/en/read"

    # Initialize Chrome webdriver
    options = webdriver.ChromeOptions()
    options.add_argument("--headless")  # Run Chrome in headless mode
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    driver = webdriver.Chrome(options=options)

    # Load the webpage
    driver.get(url)

    # Wait for some time for the page to load
    time.sleep(10)

    # Get the page source after content has loaded
    page_source = driver.page_source

    # Print the page source
    print(page_source)

    # Parse the HTML content
    soup = BeautifulSoup(page_source, 'html.parser')

    # Find all the cards containing news articles
    cards = soup.find_all(class_='news-card')

    # Check if any cards were found
    print("Number of cards found:", len(cards))

    # Initialize lists to store data
    headlines = []
    summaries = []
    links = []

    # Loop through each card and extract relevant information
    for card in cards:
        # Extract headline
        headline = card.find(class_='news-card-title').find('a').text.strip()
        headlines.append(headline)

        # Extract summary
        summary = card.find(class_='news-card-content').find('div').text.strip()
        summaries.append(summary)

        # Extract link
        link = "https://inshorts.com" + card.find(class_='news-card-title').find('a')['href']
        links.append(link)

    # Write the data to a CSV file
    with open('inshorts_articles.csv', 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(["Headline", "Summary", "Link"])
        writer.writerows(zip(headlines, summaries, links))

    print("Scraping completed and data saved to 'inshorts_articles.csv'.")

    # Quit the webdriver
    driver.quit()

if __name__ == "__main__":
    scrape_inshorts()
```
